```python
link='https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page=requests.get(link)
```

```python
page
```

```python
soup=bs(page.content,'html.parser')
print(soup.prettify())
```

```python
names=soup.find_all('span',class_='a-profile-name')
names
```

```python
cust_name=[]
for i in range(len(names)):
    cust_name.append(names[i].get_text())
cust_name
```

```python
rate=soup.find_all('span',class_='a-icon-alt')
rate
```

```python
rating=[]
for i in range(len(rate)):
    rating.append(rate[i].get_text())
rating
```

```python
title=soup.find_all('a',{"data-hook":"review-title"})
title
```

```python
titles=[]
for i in range(len(title)):
    titles.append(title[i].get_text())
titles
```

```python
titles[:]=[title.strip('\n') for title in titles]
titles
```

```python
content=soup.find_all('span',{"data-hook":"review-body"})
content
```

```python
contents=[]
for i in range(len(content)):
    contents.append(content[i].get_text())
contents
```

```python
contents[:]=[content.strip('\n') for content in contents]
contents
```
