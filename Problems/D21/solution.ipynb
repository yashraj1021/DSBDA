```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('../adult.csv')
df.head()
```

```python
df.shape
```

```python
df['Salary']=df['Salary'].replace(' <=50K', '0')
df['Salary']=df['Salary'].replace(' >50K', '1')
```

```python
plt.hist(df['Age'],bins=10,edgecolor='black')
```

```python
plt.scatter(df['capital-gain'],df['Age'],marker='o')
plt.show()
```

```python
plt.bar(df['Age'],df['capital-loss'],color='red')
```

```python
plt.plot(df['Age'],df['capital-loss'])
```

```python
level={'0':0,'1':1}
mapp=[level[numm] for numm in df['Salary']]

plt.scatter(df['Age'],df['Salary'],c=mapp,cmap='viridis')
cbar=plt.colorbar()
cbar.set_label('level')
```
